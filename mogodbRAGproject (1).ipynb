```python
import numpy as np
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document

# Step 1: Create sample documents in memory
documents = [
    Document(page_content="MongoDB is a NoSQL database that stores data in JSON-like documents."),
    Document(page_content="A collection in MongoDB is a group of documents."),
    Document(page_content="MongoDB started supporting multi-document transactions in version 4.0."),
    Document(page_content="LangChain is a framework to build applications with LLMs and vector databases."),
    Document(page_content="This is a test document without code."),
]

# Step 2: Split documents into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=50, chunk_overlap=10)
split_docs = text_splitter.split_documents(documents)
print(f"✅ Split documents into {len(split_docs)} chunks")

# Step 3: Simulate embeddings for each chunk
embedding_size = 1536
embedded_vectors = [np.random.rand(embedding_size).tolist() for _ in split_docs]
print(f"✅ Created mock embeddings for {len(embedded_vectors)} chunks")

# Step 4: Simulate a simple similarity search function
def query_data(query, top_k=3):
    # Mock embedding for query
    query_embedding = np.random.rand(embedding_size)
    
    # Compute cosine similarity
    def cosine_similarity(a, b):
        return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
    
    similarities = [cosine_similarity(query_embedding, vec) for vec in embedded_vectors]
    
    # Get top_k most similar chunks
    top_indices = np.argsort(similarities)[-top_k:][::-1]
    
    print(f"\nQuery: {query}\nTop {top_k} results:")
    for idx in top_indices:
        print(f"- {split_docs[idx].page_content} (score: {similarities[idx]:.3f})")

# Step 5: Test the mock retrieval
query_data("What is the difference between a database and a collection in MongoDB?")
query_data("When did MongoDB begin supporting multi-document transactions?")
```

```python
import numpy as np
from langchain.schema import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Sample documents with metadata
documents = [
    Document(page_content="MongoDB stores data in JSON-like documents.", metadata={"hasCode": False}),
    Document(page_content="Collections are groups of documents in MongoDB.", metadata={"hasCode": False}),
    Document(page_content="Sample code to connect MongoDB in Python.", metadata={"hasCode": True}),
    Document(page_content="LangChain helps build applications with embeddings.", metadata={"hasCode": False}),
]

# Split documents into chunks
split_docs = []
text_splitter = RecursiveCharacterTextSplitter(chunk_size=50, chunk_overlap=10)
for doc in documents:
    chunks = text_splitter.split_documents([doc])
    for c in chunks:
        c.metadata = doc.metadata  # preserve metadata
        split_docs.append(c)

# Generate mock embeddings
embedding_size = 1536
embedded_vectors = [np.random.rand(embedding_size) for _ in split_docs]

# Simple cosine similarity function
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# Retrieval function with metadata filter
def query_data(query, top_k=3, filter_metadata=None):
    query_embedding = np.random.rand(embedding_size)
    
    scores = []
    for doc, vec in zip(split_docs, embedded_vectors):
        if filter_metadata:
            skip = False
            for k, v in filter_metadata.items():
                if doc.metadata.get(k) != v:
                    skip = True
                    break
            if skip:
                continue
        scores.append((doc, cosine_similarity(query_embedding, vec)))
    
    # Sort by score
    top_results = sorted(scores, key=lambda x: x[1], reverse=True)[:top_k]
    
    print(f"\nQuery: {query}\nTop {len(top_results)} results:")
    for doc, score in top_results:
        print(f"- {doc.page_content} (score: {score:.3f})")

# Test
query_data("Find MongoDB collections", filter_metadata={"hasCode": False})
```

```python
# Step 1: Sample knowledge base
kb = [
    "MongoDB is a NoSQL database.",
    "Collections store multiple documents.",
    "Replica sets provide redundancy.",
    "Transactions in MongoDB allow multiple operations to be atomic."
]

# Step 2: Split into sentences
docs = [Document(page_content=s) for s in kb]

# Step 3: Simulate embeddings
embedding_size = 1536
embedded_vectors = [np.random.rand(embedding_size) for _ in docs]

# Step 4: Query function
def simple_qa(query, top_k=2):
    query_vec = np.random.rand(embedding_size)
    scores = [(doc.page_content, np.dot(query_vec, vec) / (np.linalg.norm(query_vec)*np.linalg.norm(vec))) 
              for doc, vec in zip(docs, embedded_vectors)]
    top_results = sorted(scores, key=lambda x: x[1], reverse=True)[:top_k]
    
    print(f"\nQuery: {query}\nTop {top_k} results:")
    for content, score in top_results:
        print(f"- {content} (score: {score:.3f})")

# Test
simple_qa("What is a replica set?")
```

```python
# Required libraries
import numpy as np
import matplotlib.pyplot as plt
from langchain.schema import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
import seaborn as sns

# ---------------------------
# Step 1: Sample knowledge base
# ---------------------------
documents = [
    Document(page_content="MongoDB is a NoSQL database.", metadata={"hasCode": False}),
    Document(page_content="Collections store multiple documents in MongoDB.", metadata={"hasCode": False}),
    Document(page_content="MongoDB replica sets provide redundancy.", metadata={"hasCode": False}),
    Document(page_content="Sample Python code to connect MongoDB.", metadata={"hasCode": True}),
    Document(page_content="LangChain enables RAG pipelines and vector memory.", metadata={"hasCode": False}),
]

# ---------------------------
# Step 2: Split text into chunks
# ---------------------------
split_docs = []
text_splitter = RecursiveCharacterTextSplitter(chunk_size=50, chunk_overlap=10)
for doc in documents:
    chunks = text_splitter.split_documents([doc])
    for c in chunks:
        c.metadata = doc.metadata
        split_docs.append(c)

# ---------------------------
# Step 3: Generate mock embeddings
# ---------------------------
embedding_size = 128  # smaller for demo
embedded_vectors = [np.random.rand(embedding_size) for _ in split_docs]

# ---------------------------
# Step 4: Cosine similarity function
# ---------------------------
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

# ---------------------------
# Step 5: Query function with metadata filtering
# ---------------------------
def query_data(query, top_k=3, filter_metadata=None):
    query_embedding = np.random.rand(embedding_size)
    scores = []
    
    for doc, vec in zip(split_docs, embedded_vectors):
        if filter_metadata:
            skip = False
            for k, v in filter_metadata.items():
                if doc.metadata.get(k) != v:
                    skip = True
                    break
            if skip:
                continue
        scores.append((doc, cosine_similarity(query_embedding, vec)))
    
    top_results = sorted(scores, key=lambda x: x[1], reverse=True)[:top_k]
    
    print(f"\nQuery: {query}\nTop {len(top_results)} results:")
    for doc, score in top_results:
        print(f"- {doc.page_content} (score: {score:.3f})")
    
    return [score for _, score in top_results]

# ---------------------------
# Step 6: Test the query
# ---------------------------
similarities = query_data("What is a MongoDB collection?", filter_metadata={"hasCode": False})

# ---------------------------
# Step 7: Visualize similarities (Heatmap)
# ---------------------------
plt.figure(figsize=(8, 4))
sns.heatmap([similarities], annot=True, cmap="YlGnBu", cbar=True)
plt.title("Top-k Similarity Scores for Query")
plt.xlabel("Document Rank")
plt.ylabel("Query")
plt.show()
```
